# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# reflect an existing database into a new model
engine = create_engine('sqlite:///hawaii.sqlite')
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 

#Starting from the last data point in the database. 
start_date = dt.date(2017,8,23)

# Calculate the date one year from the last date in data set.
prev_year = start_date - dt.timedelta(days=365)

# Perform a query to retrieve the data and precipitation scores
results = session.query(Measurement.date,Measurement.prcp).filter(Measurement.date >= prev_year).all()

# Save the query results as a Pandas DataFrame and set the index to the date column
results_df =pd.DataFrame(results, columns = ['date','precipitation'])
results_df.set_index('date',inplace=True)
# Sort the dataframe by date
results_df.sort_values('date',ascending=False,inplace=True)
# Use Pandas Plotting with Matplotlib to plot the data
results_df.plot(figsize=(10,7),rot=45,title='Daily Precipitation: 8/23/16 - 8/23/17',xlabel='Date',ylabel='Precipitation (inches)',fontsize=10).invert_xaxis()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
results_df.describe()
```

```python
# What are the most active stations?
# List the stations and the counts in descending order.
session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
session.query(Measurement.station, func.min(Measurement.tobs),func.max(Measurement.tobs),func.avg(Measurement.tobs)).group_by(Measurement.station).filter(Measurement.station == 'USC00519281').all()
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temps = session.query(Measurement.tobs).filter(Measurement.station == 'USC00519281').filter(Measurement.date >= prev_year).all()
temps_df = pd.DataFrame(temps, columns = ['Temp (F)'])
temps_df.plot(kind='hist',bins=20)
plt.tight_layout()
```

```python
# Write a function called `calc_temps` that will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start, end):
    results = session.query(func.min(Measurement.tobs),func.avg(Measurement.tobs),func.max(Measurement.tobs)).filter(Measurement.date >= start).filter(Measurement.date <= end).all()
    return results
```

```python
calc_temps('2016-09-01','2017-01-01')
```

```python
def month_temps(month):
    mo = dt.strptime(month,'%B').month
    results = session.query(Measurement.tobs).filter(func.extract('month',Measurement.date) == mo).all()
    results_df = pd.DataFrame(results,columns=[f'{month}'])
    return results_df.describe()
```
